```python
cap = cv2.VideoCapture('ScreenRecording_01-08-2019 23-21-21.MP4')
```

```python
ret, frame = cap.read()
```

```python
plt.imshow(frame)
```

```python
cap.release()
```

```python
while(cap.isOpened()):
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
